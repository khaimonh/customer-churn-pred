```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['Balance'], bins=50, edgecolor='black')
plt.title('Distribution of Account Balance')
plt.xlabel('Balance')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
